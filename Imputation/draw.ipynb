```python
#This file is to draw the compared graph

import matplotlib.pyplot as plt

name_list = ['Transition Score','Velocity Coherence','Direction Correctness','In-cluster Coherence','In-cluster Confidence']

num_list_VeloAE = [0.006302,0.393450,0.422727, 0.854273, 0.836136]
num_list_Original = [0.000764, 0.442002, 0.914171, 0.933877, 0.642019]
num_list_scVelo = [0,0.501184,0.946562,0.912494,0.884783]

sp_list_Smurf = [0.051074,0.984287,0.905779,0.999172,0.519506]
#第一个,第三个是负数
sp_list_Magic = [0.006737,0.810105,0.819739, 0.990100,0.637264]
#第一个，第三个是负数

count_list_Smurf = [0.004266,0.678708,0.316510,0.876629,0.448094]

count_list_Magic = [0.005482,0.824391,0.847966,0.987792,0.631270]

velo_list_Sumrf = [0.006027, 0.989488, 0.037566, 0.804059, 0.419416]
#第一个是负数
velo_list_Magic = [0.001147, 0.312041, 0.667056, 0.985342, 0.640385]
#第一个，第三个是负数
```

```python
x =list(range(len(name_list)))

total_width, n = 0.8, 5

width = total_width / n

fig = plt.figure(figsize = (15,8))
plt.grid(True)


plt.bar(x, num_list_scVelo, width=width, label='scVelo')
for i in range(len(x)):
    x[i] = x[i] + width

plt.bar(x, num_list_VeloAE, width=width, label='VeloAE',fc = 'y')
for i in range(len(x)):
    x[i] = x[i] + width
    
plt.bar(x, num_list_Original, width=width, label='original',tick_label = name_list,fc = 'r')
for i in range(len(x)):
    x[i] = x[i] + width

plt.bar(x, sp_list_Magic, width=width, label='Magic',tick_label = name_list,fc = 'g')
for i in range(len(x)):
    x[i] = x[i] + width

plt.bar(x, sp_list_Smurf, width=width, label='Smurf',tick_label = name_list,fc = 'b')


plt.legend()
plt.show()
```
